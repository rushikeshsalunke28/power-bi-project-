# Summary

This project involved cleaning and preparing a Netflix titles dataset for analysis. Key tasks included handling missing values in columns like director, cast, and date_added, formatting date fields, and removing extra spaces from text. Using Python libraries like pandas and numpy, the data was made consistent and ready for deeper analysis or modeling.

```python
import pandas as pd
```

```python
df=pd.read_csv("netflix_titles.csv")
```

```python
df.info()
```

# Handling and filling missing values 

```python
df.isnull().sum()
```

```python
df["director"].fillna(df["director"].mode()[0], inplace = True)
df.isnull().sum()
```

```python
df["cast"].fillna(df["cast"].mode()[0], inplace = True)
df.isnull().sum()
```

```python
df["country"].fillna(df["country"].mode()[0], inplace = True)
df.isnull().sum()
```

```python
df["date_added"].fillna(df["date_added"].mode(), inplace = True)
df.isnull().sum()
```

```python
df["rating"].fillna(df["rating"].mode()[0], inplace = True)
df.isnull().sum()
```

```python
df["duration"].fillna(df["duration"].mode()[0], inplace = True)
df.isnull().sum()
```

# Removing duplicates 

```python
df.drop_duplicates()
```

# Standardize Text Values 

```python
df['country'] = df['country'].str.capitalize().str.strip()
df
```

# Rename coumn headers to be uniform

```python
df.columns = df.columns.str.strip().str.capitalize().str.replace(" ", "_")
df
```
